```python
import pandas as pd
import psycopg
import matplotlib.pyplot as plt
```

```python
conn = psycopg.connect("dbname=pulsegrid user=justin password=password host=localhost port=5433")
```

```python
query = """
SELECT event, COUNT(*) AS total
FROM event
GROUP BY event
ORDER BY total DESC;
"""
```

```python
df = pd.read_sql_query(query, conn)
df.plot(kind="bar", x="event", y="total", legend=False)
plt.title("Event Counts by Type")
plt.show()
```
